```python
data = pd.read_csv('/content/drive/My Drive/CDAC_Datasets/Sales.csv')
data
```

```python
# new Revenue column
data['Revenue'] = data['Price'] * data['QuantitySold']
data['Revenue']
```

```python
#remove outliers in the Price column using Z-score method
def remove_outliers(df, column, threshold=3):
    z_scores = np.abs(stats.zscore(df[column]))
    return df[z_scores < threshold]

data_no_outliers = remove_outliers(data, 'Price')
data_no_outliers
```

```python
# Normalize the Price and Revenue columns using Min-Max scaling
def min_max_normalize(series):
    return (series - series.min()) / (series.max() - series.min())

data_no_outliers['Price_Normalized'] = min_max_normalize(data_no_outliers['Price'])
data_no_outliers['Revenue_Normalized'] = min_max_normalize(data_no_outliers['Revenue'])
print(data_no_outliers['Price_Normalized'])
print(data_no_outliers['Revenue_Normalized'])
```
