```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df= pd.read_csv('/content/Housing (1).csv')
print(df)
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(df['area'], df['price'], alpha=0.7, color="yellow" , edgecolors='r')
plt.title('Scatter Plot of Price vs Area')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.grid(True)
plt.show()
```

```python
avg_price_by_bedrooms = df.groupby('bedrooms')['price'].mean().sort_index()
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='bedrooms', y='price', estimator='mean',  ci=None, palette='pastel')
plt.title('Average Price by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Average Price ($)')
plt.show()
```

```python
guestroom_counts = df['guestroom'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(guestroom_counts, labels=guestroom_counts.index, autopct='%1.1f%%', startangle=140, colors=['orange', 'red'])
plt.title('Proportion of Properties with Guestroom')
plt.axis('equal')  # Equal aspect ratio ensures the pie is circular
plt.show()
```

```python
avg_price_by_bedrooms = df.groupby('bedrooms')['price'].mean().sort_index()


plt.figure(figsize=(10, 6))
plt.plot(avg_price_by_bedrooms.index, avg_price_by_bedrooms.values, marker='*', linestyle='-', color='r')
plt.title('Average Price by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Average Price ($)')
plt.grid(True)
plt.xticks(avg_price_by_bedrooms.index)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=50, color='purple', edgecolor='darkblue')
plt.title('Distribution of Property Prices')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['area'], bins=30, kde=True, color='green', edgecolor='darkgreen')
plt.title('Distribution of Property Areas')
plt.xlabel('Area (sq ft)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='bedrooms', y='price', data=df, palette='muted')
plt.title('Price Distribution by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.grid(axis='y')
plt.show()
```
